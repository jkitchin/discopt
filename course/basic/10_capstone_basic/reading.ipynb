# Lesson 10 — Capstone (Basic)

You're going to formulate, solve, and analyse a real optimization problem.
This integrates Lessons 1–9.

## Choose ONE of three problems

### A) Hospital nurse scheduling (MILP)

12 nurses, 21 shifts (3 shifts/day × 7 days). Each shift requires ≥ 3 nurses.
Each nurse can work at most 5 of 7 days, has a max of 40 hours/week, and may
not work two consecutive overnight shifts. Minimize weighted overtime cost.
Build a synthetic dataset (nurse skill levels, preferences) and report:

- the number of variables and constraints;
- the LP relaxation value;
- the MILP optimum, and the integrality gap;
- a discussion of one symmetry you can break.

### B) Refinery blending (NLP/MINLP)

5 crude streams blend into 3 product streams (gasoline, diesel, jet fuel).
Quality constraints are *bilinear* in the volumes and qualities. Maximize
profit subject to demand constraints and quality specs. Use {cite:t}`Smith1999`
or any introductory chemical-engineering reference for the formulation.
Report local vs global optima.

### C) Portfolio optimization (QP/MIQCP)

Markowitz mean-variance {cite:p}`Markowitz1952` with at most 8 assets out of
50 (cardinality constraint). Minimize variance for a target return.
Compare:
- continuous QP (no cardinality);
- MIQCP with cardinality;
- diversification penalty as a soft alternative.

## What to deliver

A single Jupyter notebook (`writing_response.ipynb`) and a short writeup
(`writing_response.md`) that includes:

1. Mathematical formulation in standard form (variables, objective, constraints).
2. The `discopt` model code, fully runnable.
3. Solver run with logs.
4. **Sensitivity analysis:** pick one parameter, perturb it, plot.
5. A reflection on what was hard, what surprised you, and what you'd do
   differently next time.

## Scope guidance

Aim for **8–12 hours of work**. Don't pick the nurse-scheduling problem if
you're new to scheduling — pick whichever maps closest to your domain.
