# Capstone — assessor notes

There is no canonical 'solution' for the capstone — students choose one of three problems. Reference formulations sketched below for the assessor.

**A) Nurse scheduling.** Variables: $x_{nst} \in \{0,1\}$ (nurse n, shift s, day t). Constraints: coverage, max-days, max-hours, no-consecutive-overnight. Symmetry: nurses with identical skill profiles → lex-order them.

**B) Refinery blending.** Variables: blend ratios + product volumes. Quality: $\sum_s q_s x_s \ge q^{\min} \sum_s x_s$ (linear) — but density-volume yields *bilinear* if both are unknown. Use McCormick relaxation (Lesson 16).

**C) Portfolio.** Min $w^\top \Sigma w$ s.t. $\mu^\top w \ge r^{\min}, \mathbb{1}^\top w = 1, \sum_i z_i \le 8, w_i \le z_i$ (cardinality). Soft alternative: penalty $\lambda \|w\|_1$.
