# Lesson 9 solutions

**E1.** KKT: $2 x_i + \lambda - \nu_i = 0, x_1 + x_2 \ge 1, x_i \ge 0, \lambda, \nu_i \ge 0,$ complementary slackness. Interior solution: $\nu = 0, x_1 = x_2 = 1/2, \lambda = -1$ violates $\lambda \ge 0$ — wait. Sign convention check: for $g_i(x) \le 0$ form, we have $-(x_1+x_2) + 1 \le 0$, multiplier $\lambda \ge 0$, gradient $(-1, -1)$, so $2x_i - \lambda = 0$, $\lambda = 2 x_i$, $x_1 = x_2 = 1/2, \lambda = 1$. Optimum $f^\star = 1/2$.
