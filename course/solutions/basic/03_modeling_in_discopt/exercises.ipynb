```python
# Add slack s_i = supply_i - sum_j x[i,j], penalty 2 * sum(s_i)
```

```python
import numpy as np, discopt as do
rng = np.random.default_rng(0); n = 30
t = np.linspace(0, 10, n); y = 2*t + 1 + rng.normal(0, 0.5, n); y[15] += 20  # outlier
m = do.Model(); a = m.add_variable(); b = m.add_variable()
u = m.add_variables(n, lb=0)
for i in range(n):
    m.add_constraint(u[i] >=  (y[i] - (a*t[i] + b)))
    m.add_constraint(u[i] >= -(y[i] - (a*t[i] + b)))
m.minimize(u.sum()); r = m.solve()
print("L1 fit a,b =", r.value(a), r.value(b))
```

```python
# A and B -> C  iff  A + B - 1 <= C, i.e., C >= A + B - 1.
```

Tight bounds reduce nodes by orders of magnitude on bilinear problems — the McCormick envelope is much tighter.
