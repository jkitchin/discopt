**E5.** $f(x,y) = x^2 - y^2$ is unbounded below — there is no minimum, only a saddle at the origin and arbitrarily-small values along the $y$-axis. GD with a non-zero $y_0$ takes ever-larger steps in $-\nabla_y f = 2y$ and **diverges to $-\infty$**; that's not 'escaping to a minimum,' it's revealing that the problem has no minimum. Newton: the Hessian $\text{diag}(2, -2)$ is indefinite, so the un-modified Newton step is not a descent direction; Newton stalls (or oscillates) without a trust-region or modified-Hessian fix. BFGS: depends on $B_0$ — if started PSD it behaves like quasi-Newton with PD curvature and resists the saddle, but the fundamental issue (no min) is unchanged.
