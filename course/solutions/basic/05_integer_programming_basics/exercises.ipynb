# Lesson 5 solutions

**E1.** $\min_x \sum a_j x_j = \sum_{a_j < 0} a_j U_j$. Take $M = b - \min_x \sum a_j x_j$. Constraint: $\sum a_j x_j + M(1-z) \ge b$, i.e., $\sum a_j x_j \ge b - M(1-z)$. With $a=(2,-1,3), U=(5,4,6), b=4$: min sum is $-4$; $M = 4 - (-4) = 8$. So constraint becomes $2x_1 - x_2 + 3 x_3 \ge 4 - 8(1-z)$.

**E5.** Issues: (1) $1e9$ is enormous; LP relaxation gives a useless bound. Derive tight $M$ from variable bounds. (2) The disjunction is not properly enforced if both branches of $z$ allow either inequality to be ignored without complementary tightening. Use convex hull or proper indicator constraints.
