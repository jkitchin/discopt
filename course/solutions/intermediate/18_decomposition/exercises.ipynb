# Lesson 18 — outlines

**E5.** Subgradient: $\lambda_{k+1} = \lambda_k + t_k g_k$ where $g_k$ is a subgradient. Step rule choices govern convergence; non-smooth objective → no monotone descent. Bundle methods aggregate past subgradients into a piecewise-linear model and use a trust region — much faster but more complex.
