# Lesson 25 — reference solutions

**E1.** For 5-variable LPs, box uncertainty typically yields the most conservative robust optimum (worst objective); ellipsoid (volume-matched, so smaller in $L^\infty$ but larger in $L^2$) sits between box and nominal; ellipsoid solutions often have *more nonzero entries* than box (which tends to push to extreme vertices).

**E2.** Price-of-robustness curve: as $\Gamma$ goes 0 → n, expected return strictly decreases while worst-case return strictly increases; the curve is concave by construction. Typical behaviour on a 5-asset portfolio: $\Gamma = 1$ already captures ~70 % of the worst-case improvement of $\Gamma = n$, justifying small $\Gamma$ in practice.

**E3.** Affine LDR-RO improves on static RO by 5-20 % in expected cost on inventory problems with bounded demand; the improvement grows with problem horizon.

**E4.** RO with $\Gamma = 3$ vs SP with 100 scenarios: RO's $x$ is typically more conservative (extreme on a few coordinates) while SP averages across scenarios; RO is more robust to distribution mis-specification, SP is better calibrated to the chosen distribution.

**E5.** Robust scheduling: budget formulation with $\Gamma = $ (#tasks)/3 captures the bulk of worst-case protection at modest expected-makespan cost.
