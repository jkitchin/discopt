**E1.** $\arg\min_x (x - \theta)^2 + 0.1 x^2$ has $x^\star = \theta / 1.1$, so $dx^\star/d\theta = 1/1.1 \approx 0.909$. Implicit-function: $\nabla_x f = 2(x - \theta) + 0.2 x$, $\nabla_{xx}^2 f = 2.2$, $\nabla_{x\theta}^2 f = -2$, so $dx^\star/d\theta = -(-2)/2.2 = 0.909$. ✓

**E2.** A correct cvxpylayers QP with $\Sigma$, equality $\mathbb{1}^\top w = 1$, bound $w \ge 0$ should match a finite-difference gradient (perturb $\mu$ by 1e-5) to ~1e-4 relative.

**E3.** Decision-focused training generally gives 5-30 % lower realized regret than MSE training when (a) the underlying signal-to-noise ratio is moderate and (b) the optimization decision is sensitive to a *direction* of the prediction more than its magnitude. With high SNR, MSE ≈ DF.

**E4.** LP differentiation breaks at *vertex transitions* (degeneracy): the argmin jumps discontinuously when the cost vector crosses a normal-cone boundary. Within a single basis the gradient is well-defined and equals $-(B^{-1})$ row of the binding constraint.

**E5.** Soft penalty is preferable when (a) the constraint is inherently approximate (e.g., 'roughly' centered), (b) the QP layer would be the bottleneck of the forward/backward pass, or (c) the constraint set has empty interior at certain inputs (Slater fails).
