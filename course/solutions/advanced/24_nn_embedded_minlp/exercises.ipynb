**E1.** Branin's true maximum on $[-5,10] \times [0,15]$ is $\approx 308.13$ near the boundary. A 2-hidden-layer ReLU MLP (e.g., 16-16) trained on 200 LHS samples typically attains test-set RMSE around 5–15 % of the function's range. Embedding via `discopt.nn` and running `mode='global'` reports the surrogate's true maximum, which usually differs from Branin's true max by the surrogate error.

**E2.** With $\underline a = -2, \overline a = 3$ on a single ReLU, the four constraints are $z \ge a, z \ge 0, z \le a - (-2)(1-y) = a + 2(1-y), z \le 3y$, $y \in \{0,1\}$. Substituting against `discopt.nn`'s output should agree to 1e-9 on a network with the same weights.

**E3.** OBBT layer-by-layer typically gives 1.5–4x speedup on networks with 2-3 hidden layers; the speedup grows with network depth because IBP bounds compound exponentially.

**E4.** Replacing ReLU with tanh produces a smooth nonconvex MINLP — `mode='global'` runtimes typically jump 10–100x because spatial branching now operates on continuous nonlinear curvature rather than discrete activation choices.

**E5.** A correct surrogate-loop iteration: (i) sample N points uniformly, train NN; (ii) embed-and-solve to get candidate $x^*$; (iii) evaluate true simulator at $x^*$, append to dataset, retrain. On a smooth function, the candidate typically beats the random max by 1-2 percentile points after a single iteration.
