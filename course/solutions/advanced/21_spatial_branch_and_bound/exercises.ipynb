# Lesson 21 — reference solutions

**E1.** $z = xy$ on $[0,2]^2$. McCormick at root: $z \ge \max(0, 2x + 2y - 4)$, $z \le 2 \min(x, y)$. Active LP minimum on $x + y \ge 1$: any of $(1, 0), (0, 1)$ → $z_{LP} = 0$ (loose). Branch $x$ at 1: child A ($x \in [0, 1]$) gives $z_{LP}^A = 0$; child B ($x \in [1, 2]$) forces $y \ge 0$ and $z \ge 0$ — but in B the constraint $x + y \ge 1$ is automatic so $y$ can be 0, $z = 0$. Both children meet $z = 0$, which is the global optimum (achieved at $(1, 0)$, $(0, 1)$, etc.).

**E2.** Typical observation on a 3-bilinear MINLP: OBBT shrinks the root box by 30–60 % and reduces node count by 3–10x. The OBBT sub-LPs add overhead but are amortised across the tree.

**E3.** `most_violated` typically beats `longest_edge` on bilinear problems by 1.5–3x in node count when many auxiliaries are present; `longest_edge` can win on highly anisotropic boxes.

**E4.** Minimal spatial-B&B sketch: recurse on box; at each node, build LP with McCormick auxiliaries via `scipy.optimize.linprog`, branch on midpoint of the most-violated auxiliary; track `(LB, UB, incumbent)` globally.

**E5.** Diagnostic checklist: (i) examine bound widths — if any auxiliary box stays $\ge 50\%$ of the original after OBBT, bounds are the bottleneck; (ii) check for symmetric variables in the formulation; (iii) measure McCormick gap at the LP solution — large gap = need RLT or piecewise McCormick.
